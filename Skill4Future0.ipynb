```python
data1['Month'] = pd.to_datetime(data1['Date']).dt.month

average_orders_per_state_monthly = data1.groupby(['ship-state', 'Month'])['Order ID'].count()
average=average_orders_per_state_monthly.groupby('ship-state').mean()
print(average_orders_per_state_monthly)
print(average)
```

```python
# Assuming you have 'Amount' and 'Year' columns
data1['Year'] = pd.to_datetime(data1['Date']).dt.year  # Extract year

average_sales_per_state_yearly = data1.groupby(['ship-state', 'Year'])['Amount'].sum().groupby('ship-state').mean()

# Plotting
average_sales_per_state_yearly.plot(kind='bar', figsize=(20, 8))
plt.title('Average Yearly Sales per State')
plt.xlabel('State')
plt.ylabel('Average Sales')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: find the co-relation between  all the available variables without errors

import pandas as pd
import matplotlib.pyplot as plt
# ... (your existing code)

# Select numerical columns for correlation analysis
numerical_cols = data1.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_cols.corr()

# Print or visualize the correlation matrix
print(correlation_matrix)

# You can also create a heatmap for better visualization
plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.index)), correlation_matrix.index)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
